# Исследование на тему: "Определение неэффективных операторов"

Заказчиком данного исследования является продуктовый департамент оператора связи "Нупозвони". Компания предоставляет услуги колл-центра любым юрлицам (в дальнейшем - клиентам). У каждого клиента есть какое-то количество операторов, которые занимаются обзвоном (входящие, исходящие либо смешанные звонки). Необходимо найти самых неэффективных операторов. Признаки низкой эффективности:

- много пропущенных входящих вызовов,
- долгое ожидание ответа при входящих внешних вызовах,
- мало исходящих вызовов — в случае тех колл-центров, что специализируются на обзвонах, а не обработке входящих звонков.

Предполагается, что выявление неэффективных операторов, позволит сервису увеличить свою ценность и срок жизни клиентов.

По результатам исследования департамент ожидает получить таблицу неэффективных операторов по показателям, подтверждающим низкую эффективность. 

**Цель исследования:**

Выработка постоянного алгоритма для определения неэффективных операторов.

**Задачи:**

- категоризация колл-центров (специализация на исходящих, входящих или смешанных звонках);
- анализ цикличной нагрузки на операторов (пиковые дни работы операторов);
- исследование деятельности операторов по признакам низкой эффективности;
- определение количества клиентов, потенциально заинтересованных в предложении услуг по результатам исследования.

**Описание данных:**

Датасеты описывают использование услуг «Нупозвони» — провайдера виртуальный телефонии. Его клиенты — колл-центры, которые: 

- распределяют входящие вызовы на операторов,
- совершают исходящие вызовы силами операторов.

Также операторы могут делать внутренние вызовы —  вызовы друг между другом внутри сети виртуальной телефонии.

Колонки в `telecom_dataset.csv.zip`:

- `user_id` — Идентификатор клиентского аккаунта в сервисе
- `date` — Дата статистики
- `direction` — Направление вызовов (out - исходящий вызов, in — входящий вызов)
- `internal` — Является ли звонок внутренним звонком между операторами клиента
- `operator_id` — Идентификатор оператора
- `is_missed_call` — Является ли звонок пропущенным
- `calls_count` — Количество звонков
- `call_duration` — Длительность звонка (без учета времени ожидания)
- `total_call_duration` — Длительность звонка (с учетом времени ожидания)

Колонки в `telecom_clients.csv`:

- `user_id` — Идентификатор клиентского аккаунта в сервисе
- `tariff_plan` — Текущий тарифный план клиента
- `date_start` — Дата регистрации клиентов в сервисе

Мы не можем идентифицировать пропущенных операторов, но и удалять их тоже не хотелось бы, чтобы не потерять 15 % всех наших данных. Заполним пропущенные значения в столбце operator_id самым часто встречающимся значением айди оператора по каждому клиенту.

```python
mode =  df['operator_id'].mode()[0]
df['operator_id'] = df.groupby('user_id')\
['operator_id'].apply(lambda x: x.fillna(x.value_counts()\
                                         .idxmax() if x.value_counts().max() >=1 else mode , inplace = False))
df.info()
```

Для дальнейшего адекватного восприятия деятельность операторов, категоризируем клиентов (колл-центры), в которых работают операторы. Выделим три категории, исходя из специализации колл-центров:

- колл-центры, работающие на приём звонков;
- колл-центры,работающие в смешанном режиме.

В процессе исследования данных, определим  тот барьер, который будем использовать в качестве категоризации колл-центров.

Создадим сводную таблицу, для категоризации.

```python
df_pivot = df.pivot_table(index=['tariff_plan','user_id', 'operator_id', 'date', 'calls_count'], columns='direction', values='total_call_duration', aggfunc='sum')\
.reset_index()
df_pivot
```

```python
df_pivot = df.pivot_table(
    index=[
        'tariff_plan',
        'user_id', 
        'operator_id', 
        'date', 
        'calls_count'
    ], 
    columns='direction',
    values='total_call_duration', 
    aggfunc='sum'
).reset_index()
df_pivot
```

Заменим nan в таблице на нули.

```python
df_pivot = df_pivot.fillna(0)
```

Определим столбцы с долями для входящих и исходящих звонков.

```python
df_pivot['call_in_perc'] = df_pivot['in'] / (df_pivot['in'] + df_pivot['out']) * 100
```

```python
df_pivot['call_out_perc'] = df_pivot['out'] / (df_pivot['in'] + df_pivot['out']) * 100
```

```python
df_pivot
```

Посмотрим на статистику распределения сумм секунд звонков среди операторов, затраченных на входящие и исходящие вызовы.

```python
df_pivot['out'].describe()
```

Представляется наиболее приемлемым произвести категоризацию по 75 % перцентилю. Правда, есть подозрение, что может образоваться большое число id, которые попадут в категорию, работающую над смешанными звонками. Данные покажут.

```python
def quantile_call(column, percent):
    count = 0
    for i in df_pivot[column]:
        if i >= percent:
            count += 1
        else:
            continue
    return count    
```

```python
quant_in = quantile_call('call_in_perc', 75)
quant_in
```

```python
quant_out = quantile_call('call_out_perc', 75)
quant_out
```

```python
mix = len(df_pivot) - quant_in - quant_out
mix
```

Подозрения не подтвердились. Поставим метки в сводной таблице для обозначения категории колл-центра.

Обозначим новые колонки:
- incoming call - входящий вызов;
- outgoing call - исходящий вызов.

```python
df_pivot['incoming call'] = df_pivot['call_in_perc'] >= 75
```

```python
df_pivot['outgoing call'] = df_pivot['call_out_perc'] >= 75
```

```python
df_pivot
```

Для удобства визуализации присвоим значениям False значение нуля, а True - единицы.

```python
df_pivot = df_pivot.replace(False, 0)
    
df_pivot = df_pivot.replace(True, 1)    
```

```python
df_pivot
```

Исхитримся для создания колонки - mixed challenge - смешанный вызов.

```python
df_pivot['mixed challenge'] = 1 - df_pivot['incoming call'] - df_pivot['outgoing call'] 
```

Определим клиентов, потенциально заинтересованных в предложении услуг. Для этого создадим две переменные: с id неэффективных операторов входящих звонков и с id неэффективных операторов исходящих звонков.

```python
inefficient_operator_id_in = inefficient_incoming_operators['operator_id']
inefficient_operator_id_out = inefficient_outgoing_operators['operator_id'].unique()
```

```python
inefficient_operator_id_in
inefficient_operator_id_out

interested_clients = (df.query('operator_id in @inefficient_operator_id_in or operator_id in @inefficient_operator_id_out')
                                   .pivot_table(index = ['user_id', 'operator_id'],
                                                values = 'calls_count', aggfunc = 'sum')
                                   .reset_index().sort_values(by = 'calls_count', ascending = False))
interested_clients
```

```python
interested_clients['user_id'].unique()
```

Итого, мы получили список из 53 клиентов, которым потенциально интересно было бы узнать список неэффективных операторов для повышения эффективности своей деятельности. Для того, чтобы наше предложение действительно имело ценность для клиентов, определим какое количество операторов имеет тот или иной клиент. Идея заключается в том, что клиент, у которого есть 1-2 оператора сможет определить уровень их эффективности и без нашей помощи, в ручном режиме. 

```python
for i in interested_clients['user_id'].unique():
    count = []
    
    for k, j in zip(interested_clients['user_id'],  interested_clients['operator_id']):
        if i == k:
            count.append(j)
    print(f'Количество неэффективных операторов клиента с id: {i}, равно {len(count)}, неэффективные операторы: {count}')
            
```

---

Ранжируем клиентов по востребованности той услуги, которую мы хотим им предоставить. Логика определения ранга предельно примитивна: чем больше операторов, тем выше ранг.

```python
rank1 = []
rank2 = []
rank3 = []

for i in interested_clients['user_id'].unique():
    count = []
    
    for k, j in zip(interested_clients['user_id'],  interested_clients['operator_id']):
        if i == k:
            count.append(j)
    if len(count) <= 2:
        rank3.append(i)
    elif 2 < len(count) < 5:
        rank2.append(i)
    else:
        rank1.append(i)
        
    
print(f'Наиболее перспективные клиенты с id: {rank1}, имеющие более пяти операторов') 
print()
print(f'Перспективные клиенты с id: {rank2}, имеющие более двух операторов')
print()
print(f'Наименее перспективные клиенты с id: {rank3}, имеющие 1-2 операторов и способные справиться с ситуацией самостоятельно')
```
